```python
import pandas as pd
import openpyxl
import yfinance as yf
```

```python
goto = yf.Ticker("GOTO.JK")
goto.info['sector']
```

```python
file = './20230418.xlsx'
result = './result.csv'
```

```python
## Load File
read_file = pd.read_excel(file)
csv_data_df = read_file.to_csv(result)
csv_data_df = pd.read_csv(result)
```

```python
def assigning(df):
  print(len(df))
  industry = []
  sector = []
  for row in df['Kode Saham']:
    tick = yf.Ticker(row.upper() + '.JK')
    industry.append(tick.info['industry'])
    sector.append(tick.info['sector'])
  df['Industry'] = industry
  df['Sector'] = sector
  return df
```

```python
## Foreign Buy
df = csv_data_df[['Kode Saham', 'Foreign Buy']].sort_values(by=['Foreign Buy'], ascending=False)
df = df.head(20)
# df = result.reset_index()
# df

new_df = assigning(df)
new_df
```

```python
## Foreign Sell
df = csv_data_df[['Kode Saham', 'Foreign Sell']].sort_values(by=['Foreign Sell'], ascending=False)
df = df.head(20)

new_df = assigning(df)
new_df
```

```python
## Net Buy
df2 = csv_data_df.assign(NetBuy = lambda row: row['Foreign Buy'] - row['Foreign Sell'])
df = df2[['Kode Saham', 'Foreign Buy', 'Foreign Sell', 'NetBuy']].sort_values(by=['NetBuy'], ascending=False)
df = df.head(20)

new_df = assigning(df)
new_df
```

```python
## Net Sell
df2 = csv_data_df.assign(NetSell = lambda row: row['Foreign Sell'] - row['Foreign Buy'])
df = df2[['Kode Saham', 'Foreign Buy', 'Foreign Sell', 'NetSell']].sort_values(by=['NetSell'], ascending=False)
df = df.head(20)

new_df = assigning(df)
new_df
```
